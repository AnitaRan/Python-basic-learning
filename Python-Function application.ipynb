## 1. 获取每个月的最高温度

```python
df = pd.read_csv('./data/BJ_tianqi.csv')

# 将日期的列转换成时间类型
df['日期'] = pd.to_datetime(df['日期'])

# 增加‘月份’列
df['月份'] = df['日期'].dt.to_period('M')

# 温度中的℃去掉，数据转化成int
df['最高温度'] = [int(i.replace('℃','')) for i in df['最高温度']]
df['最低温度'] = [int(i.replace('℃','')) for i in df['最低温度']]

# 将‘日期’列转换成行索引
df.set_index('日期',inplace=True)

df
```

```python
df.groupby('月份')['最高温度'].max()
```

## 2. 获取每个月的最高温和最低温度

```python
#  给每列作用不同的聚合函数
dict_mapping = {
     '最高温度':'max',
    '最低温度':'min'
}

df.groupby('月份').agg(dict_mapping)
```

## 3. 取出每个月最高气温的前二

```python
def month_toptemp(month,toptemp,n):
    return month.sort_values(toptemp)[-n:]

df.groupby('月份')[['最高温度']].apply(month_toptemp,toptemp='最高温度',n=2)
```
