## 设置图片的大小：plt.figure(figsize=(20,8),dpi=80)
- figsize:指定figure的宽和高，单位为英寸(1英寸等于2.5cm,A4纸是 21*30cm的纸张)；
- dpi参数指定绘图对象的分辨率，即每英寸多少个像素，缺省值为80       

```python
import random

x = range(2,26,2)
y = [random.randint(15,30) for i in x]

plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)
plt.savefig('./output/t1.svg')
plt.show()
# 注意：保存要放在显示的前面,因为plt.show()会释放figure资源，如果在显示图像之后保存图片将只能保存空图片。
# 图片的格式也可以保存为svg这种矢量图格式，这种矢量图放在网页中放大后不会有锯齿
```

## 设置刻度xticks
- 第一个参数确立了疏密程度，第二个参数设置了刻度显示的内容

```python
x = range(2,26,2)
y = [random.randint(15,30) for i in x]

plt.figure(figsize=(20,8),dpi=80)

# 设置x轴的刻度：xticks 第一个参数确立了疏密程度，x轴有多少个刻度
# 构造x轴刻度标签：x_ticks_label 真正显示的内容
#如果没有设置刻度显示的内容,默认显示的内容就是你传递的列表元素的内容

# plt.xticks(range(1,25))
# plt.yticks(range(min(y),max(y)+1))

# rotation 旋转度数
x_ticks_label = ['{}:00'.format(i) for i in x]
y_ticks_label = ['{}摄氏度'.format(i) for i in range(min(y),max(y)+1)]

plt.xticks(x,x_ticks_label,rotation = 45)
plt.yticks(range(min(y),max(y)+1),y_ticks_label)

plt.plot(x,y)
plt.show()
```

## 设置标题和坐标轴标签：title、xlabel、ylabel

```python
# 两个小时内的每分钟跳动变化

x = range(0,120)
y = [random.randint(10,30) for i in x]

plt.figure(figsize=(20,8),dpi=80)
plt.plot(x,y)

plt.xlabel('时间',fontsize=15)
plt.ylabel('次数',fontsize=15)
plt.title('每分钟跳动次数',color='red',fontsize=20)

plt.show()
```

```python
# 假设大家在30岁的时候，根据自己的实际情况，统计出来你和你同事各自从15岁到30岁每年交的男/女朋友

x = range(15,31)
y1 = [1,0,1,1,2,4,3,4,4,5,6,5,4,3,3,1]
y2 = [1,0,3,1,2,2,3,4,3,2,1,2,1,1,1,1]

plt.figure(figsize=(20,8),dpi=80)


# 设置刻度xticks：第一个参数确立了疏密程度，第二个参数设置了刻度显示的内容
x_ticks_label = ['{}岁'.format(i) for i in x]
plt.xticks(x,x_ticks_label,rotation=45)

# 绘制网格grid——可以设置线的样式、透明度
plt.grid(alpha=0.5)

plt.plot(x,y1,color='red',label='自己')
plt.plot(x,y2,color='blue',label='同事')

# 显示图例legend——设置位置loc（best、upper right、upper left、lower left、lower right、right、center left、center right、lower center、upper center、center）,也可以不设置   
plt.legend(loc='upper right')

plt.show()
```

## 绘制子图：add_subplot

```python
import numpy as np

x = np.arange(1,100)

#新建figure对象,fig获取画布对象
fig = plt.figure(figsize=(20,10),dpi=80)

# 新建子图1
ax1 = fig.add_subplot(2,2,1)
ax1.plot(x,x)

# 新建子图2
ax2 = fig.add_subplot(2,2,2)
ax2.plot(x,x**2)

ax2.grid(color='r',linestyle='--',linewidth=1,alpha=0.3)

# 新建子图3
ax3=fig.add_subplot(2,2,3)
ax3.plot(x, np.log(x))

plt.show()
```

## 截取图形部分：xlim、ylim

## 移动坐标轴：set_position

```python
x = [-3,-2,-1,0,1,2,3]
y = range(0,14,2)

# 获得当前图表的图像
ax = plt.gca()

# 设置图型的包围线
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.spines['bottom'].set_color('blue')
ax.spines['left'].set_color('red')

# 设置底边的移动范围，移动到y轴的0位置,'data':移动轴的位置到交叉轴的指定坐标
ax.spines['bottom'].set_position(('data', 0))
ax.spines['left'].set_position(('data', 1))

plt.plot(x,y)
plt.show()
```
