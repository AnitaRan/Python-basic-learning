## 6.数据的合并 concat、join、merge

```python
# 数据的合并--concat：沿着一个轴进行合并

df1 = pd.DataFrame(np.arange(6).reshape(3,2),columns=['four','five'])
df2 = pd.DataFrame(np.arange(6).reshape(2,3),columns=['one','two','three'])
print(df1)
print(df2)
print('\n')

result1 = pd.concat([df1,df2],axis=0,ignore_index=True)
print(result1)
print('\n')

result2 = pd.concat([df1,df2],axis=1,ignore_index=True)
print(result2)
```

```python
# 数据的合并--join:'left', 'right', 'outer', 'inner'，默认‘left’，关注行的合并

df3 = pd.DataFrame({'Red':[1,3,5],'Green':[5,0,3]},index=list('abc'))
df4 = pd.DataFrame({'Blue':[1,9,8],'Yellow':[6,6,7]},index=list('cde'))
print(df3)
print(df4)
print('\n')

result3 = df3.join(df4,how='right')
print(result3)
print('\n')

result4 = df3.join(df4,how='outer')
print(result4)

# 合并多个对象
df5 = pd.DataFrame({'brown':[3,4,5],'white':[1,1,1]},index=list('aed'))
df3.join([df4,df5])
```

```python
# 数据的合并--merge：两个数据之间使用哪个列进行连接,关注列的合并
# 默认下是根据左右对象中出现同名的列作为连接的键，且连接方式是how='inner'

df1=pd.DataFrame({'名字':list('ABCDE'),'性别':['男','女','男','男','女'],
                  '职称':['副教授','讲师','助教','教授','助教']},
                 index=range(1001,1006))
df1.index.name = '编号'
df1.columns.name = '学院老师'
print(df1)

df2=pd.DataFrame({'名字':list('ABDAX'),'课程':['C++','计算机导论','汇编','数据结构','马克思原理'],
                  '职称':['副教授','讲师','教授','副教授','讲师']},
                 index=[1001,1002,1004,1001,3001])
df2.index.name = '编号'
df2.columns.name = '课程'
print(df2)
print('\n')

# 根据指定列名进行合并
# pd.merge(df1,df2,on='名字',suffixes=['_1','_2'])

# 根据多个键进行合并
pd.merge(df1,df2,on=["名字","职称"])

# pd.merge(df1,df2,how='inner')
# pd.merge(df1,df2,how='left')
# pd.merge(df1,df2,how='right')
# pd.merge(df1,df2,how='outer')
```

## 8.空值处理 dropna

```python
from numpy import nan as NaN

# 处理Series对象
se = pd.Series([4,NaN,8,NaN,5])
# print(se)
# print(se.isnull())
# print(se.notnull())
# print(se.dropna())

# 处理DataFrame对象
df = pd.DataFrame([[1,2,3],[NaN,NaN,2],[NaN,NaN,NaN],[8,8,NaN]])
print(df)

# dropna()默认滤除所有包含NaN
print(df.dropna())

# 传入 how='all' 滤除全为NaN的行
print(df.dropna(how='all'))

# 传入 axis=1 滤除列(不传入，默认axis=0)
print(df.dropna(how='all',axis=1))

# 传入 thresh=n 保留至少有n个非NaN数据的行
print(df.dropna(thresh=1))
```

## 10.分组聚合

```python
df=pd.DataFrame({
    'name':['BOSS','Lilei','Lilei','Han','BOSS','BOSS','Han','BOSS'],
    'Year':[2016,2016,2016,2016,2017,2017,2017,2017],
    'Salary':[999999,20000,25000,3000,9999999,999999,3500,999999],
    'Bonus':[100000,20000,20000,5000,200000,300000,3000,400000]
    })

df
```

```python
group_by_name = df.groupby('name')

# 查看分组 groups
# print(group_by_name)
print(group_by_name.groups)

# 分组后的数量 count
print(group_by_name.count())

# 获取一组数据 get_group
group_boss = group_by_name.get_group('BOSS')
print(group_boss)


# 分别获取每组数据内容
for name,group in group_by_name:
    print(name)# 组的名字
    print(group)# 组具体内容
```

```python
group_by_name_year=df.groupby(['name','Year'])

print(group_by_name_year.get_group(('BOSS', 2016)))
print('\n')

for name,group in group_by_name_year:
    print(name)# 组的名字
    print(group)# 组具体内容
```

```python
# 将某列数据按数据值分成不同范围段进行分组（groupby）运算

df = pd.DataFrame({'Age': np.random.randint(20, 70, 100), 
                    'Sex': np.random.choice(['M', 'F'], 100), 
                    })

df
```

```python
age_groups = pd.cut(df['Age'], bins=[19,40,65,100])
# print(type(age_groups))

print(df.groupby(age_groups).count())

# 按‘Age’分组范围和‘Sex’进行制作交叉表
pd.crosstab(age_groups,df['Sex'])
```

## python高级函数：agg与apply
- agg 只能返回一个结果值
- apply 可以返回一系列值

```python
df=pd.DataFrame({
    'name':['BOSS','Lilei','Lilei','Han','BOSS','BOSS','Han','BOSS'],
    'Year':[2016,2016,2016,2016,2017,2017,2017,2017],
    'Salary':[999999,20000,25000,3000,9999999,999999,3500,999999],
    'Bonus':[100000,20000,20000,5000,200000,300000,3000,400000]
    })

df
```

```python
df.groupby('Year').sum()
```

```python
def peak_range(df):
    return df.max() - df.min()

df.groupby(['name','Year']).agg(['mean',peak_range])
```

```python
#  给每列作用不同的聚合函数

dict_mapping = {
    'Salary':['mean','max'],
    'Bonus':'sum'
}

df.groupby('name').agg(dict_mapping)
```

```python
# 取出抽烟和不抽烟的体重前二
# 思路：smoker 进行是否抽烟分组的,分组后的数据传递

df1=pd.DataFrame({'sex':list('FFMFMMF'),'smoker':list('YNYYNYY'),'age':[21,30,17,37,40,18,26],'weight':[120,100,132,140,94,89,123]})
print(df1)

def top(smoker,weight,n):
    return smoker.sort_values(weight)[-n:]

df1.groupby('smoker').apply(top,weight='weight',n=2)
```

## 拓展：多重索引

```python
#  Series单层索引
s1 = Series(np.random.randint(0,150,size=6),index=list('abcdef'))
print(s1)

# Series创建多层索引
s2 = Series(np.random.randint(0,150,size=6),
           index=[['a','a','b','b','c','c'],
                  ['期中','期末','期中','期末','期中','期末']])
print(s2)

# 取一个第一级索引
print(s2['a'])

# 取多个第一级索引
print(s2[['a','b']])

# 根据索引获取值
print(s2['a']['期中'])

# loc方法取值
# print(s.loc['a'])
# print(s.loc[['a','b']])
print(s2.loc['a']['期末'])
```

```python
#  DataFrame创建多层索引
df1 = DataFrame(np.random.randint(0,150,size=(6,4)),
               columns = ['zs','ls','ww','zl'],
               index = [['python','python','math','math','En','En'],
                        ['期中','期末','期中','期末','期中','期末']])
print(df1)
```

```python
# 特定结构 ：pd.MultiIndex.from_arrays

class1=['python','python','math','math','En','En']
class2=['期中','期末','期中','期末','期中','期末']

# 构造一个多层索引对象
m_index2 = pd.MultiIndex.from_arrays([class1,class2])

df1 = DataFrame(np.random.randint(0,150,(6,4)),index=m_index2)
df1
```

```python
# 特定结构：pd.MultiIndex.from_product

class1 = ['python','math','En']
class2 = ['期中','期末']

m_index2=pd.MultiIndex.from_product([class1,class2])

df2 = DataFrame(np.random.randint(0,150,(6,4)),index=m_index2)

print(df2)

# 获取列
print(df2[0])

# 一级索引
print(df2.loc['python'])

# 多个一级索引
print(df2.loc[['python','math']])

# 取一行
print(df2.loc['python','期末'])

# 取一值
print(df2.loc['python','期中'][0])

# iloc是只取最内层的索引的
print(df2.iloc[0])   
print(df2.iloc[0,1]) 
```
